We see that there are no null values. So, we are good to go with further visual analysis. The library: matplotlib can do good job of visualizing data. However, there is a better library with more functionalities: [seaborn](https://seaborn.pydata.org/). This is built on the top of matplotlib. We will be using this library for visualization. Installation is really easy: use conda (or pip) to install. *conda install seaborn* (Refer: [seaborn installation](https://seaborn.pydata.org/installing.html)). After that follow the next steps:

```python
import seaborn as sns
```

seaborn's pairplot function which is very useful to scatter plot all the features at once. We would have to otherwise keep plotting them individually.

```python
sns.pairplot(data)
```

You may choose to ignore the main diagonal. Just look at non-diagonal plots and see what kind of stories these plots are telling.
<br/>

Now let's see some correlation plots.

```python
plt.figure(figsize=(10,11))
sns.heatmap(data.corr(),annot=True)
plt.plot()
```

Interesting, isn't it? What do you see? The correlation values lies between [-1,1]: "-1" means highest negative correlated and "+1" means highest positive correlation. Negative correlation means "two series" going opposite to each other *e.g.* [1,2,3,4,5] and [6,5,4,3,2].
<br/>

petal_length and petal_width are very highly correlated: 0.96. Whereas, petal_length and sepal_width are very negatively correlated: -0.42. Diagonal elements are correlation of a variable with itself, which is always 1.

<br/>

Now, Letâ€™s see how our data is distributed based on Sepal Length and Width features using a scatterplot:

```python
sns.FacetGrid(data,hue="Class").map(plt.scatter,"sepal_length","sepal_width").add_legend()
plt.show()
```

What do you see? We are aware that there 3 classes in our dataset. Now, the scatter plot is saying that given two variables sepal_length and sepal_width, it is not easy to separate two classes such as versicolor and virginica. What else it is saying? (Answer yourself)

<br/>

Now, lets plot two other variables: Petal Length and Width

```python
sns.FacetGrid(data,hue="Class").map(plt.scatter,"petal_length","petal_width").add_legend()
plt.show()
```

Life is little easy with these two features i.e. the separation of classes is easier. Now, let's plot sepal length and petal length.

```python
sns.FacetGrid(data,hue="Class").map(plt.scatter,"sepal_length","petal_length").add_legend()
plt.show()
```

There is a plot called *violin plot*. The violinplot shows density of the length and width in the species. The thinner part denotes that there is less density whereas the fatter part conveys higher density. This is shown for each class.

```python
plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
sns.violinplot(x="Class",y="sepal_length",data=data)
plt.subplot(2,2,2)
sns.violinplot(x="Class",y="sepal_width",data=data)
plt.subplot(2,2,3)
sns.violinplot(x="Class",y="petal_length",data=data)
plt.subplot(2,2,4)
sns.violinplot(x="Class",y="petal_width",data=data)
```

And similarly use boxplot to see how the categorical feature "Class" is distributed with all other four input variables. Remember, box plot from the lecture 05 (C.H. 9,10)?

```python
plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
sns.boxplot(x="Class",y="sepal_length",data=data)
plt.subplot(2,2,2)
sns.boxplot(x="Class",y="sepal_width",data=data)
plt.subplot(2,2,3)
sns.boxplot(x="Class",y="petal_length",data=data)
plt.subplot(2,2,4)
sns.boxplot(x="Class",y="petal_width",data=data)
```

You are able to see some outliers: those black diamond type dots.
